Memory-Based To-Do List:
- The Goal: An application that lets a user add a task, view all tasks, remove a task, and exit. Store the tasks in a Python List.
- Concepts Tested: List methods (.append(), .pop()), indexing, iterating through lists with for loops.

```python
task = [] 
user_task = input("Enter ur task :")
task.append(user_task)
while True:
    user_choice = input("Enter ur choice between add, view, remove, exit :").lower()
    match user_choice:
        case "add":
            user_add = input("Enter task to be added :")
            add_order = int(input("Enter the order you want to add :"))
            user_add = user_add.strip()
            task.insert(add_order, user_add)
            task.append(user_add)
            print("The task has been added")
        case "view":
            i = int(input("Enter the number of task :"))
            if i > len(task):
                print ("Not founded")
            else:
                print ("Your task :")
                print(f"{i}.{task[i-1]}")
        case "remove":
            remove_task = input("Enter the task :")
            if remove_task not in task:
                print ("Not founded")
            else:
                task.remove(remove_task)
                print ("The task has been removed")
        case "exit":
            print ("The program end!")
            break
```

Interactive Flashcard / Quiz App:
- The Goal: Create a dictionary where the keys are questions (e.g., countries) and values are answers (e.g., capitals). Loop through the dictionary, ask the user the question, check their answer, and calculate a final score.
- Concepts Tested: Dictionaries, key-value pairs, iteration, score tracking.

```python
Aboutme = {
    "what is my name?" : "Chhayleng",
    "how old I am ?" : 19,
    "where I am from?" : "Phnom Penh",
    "what is my favorite color?" : "brown"
}
user_score = 0
for question, answer in Aboutme.items():
    ans = input(f"Question :{question}").lower()
    if ans == str(answer).lower() :
        print("you are correct!")
        user_score += 1
    else :
        print ("Incorrect!")
    print (f"Your scores : {user_score}")
    print(f"Answer : {answer}")
```

The Shopping Cart Analyzer
- The Goal: Ask the user to input the prices of items they are adding to their cart, one by one. Store these prices in a Python List. The loop stops when the user types 'done'. Then, calculate and print a receipt showing the total cost, the most expensive item in the cart, and the average item price.
- Concepts Tested: while loops for input collection, list methods (.append()), built-in list functions (sum(), max(), len()), and data type casting (handling the mix of string commands and float prices).

```python
items = []
while True :
    user_choice = input("Enter price of the item or done :").lower()
    if user_choice == "done":
        break
    else:
        price = float(user_choice)
        items.append(price)
total = sum(items)
print (f"Total amount = {total}")
max_price = max(items)
print (f"The most expensive item is = {max_price}")
average = total/len(items)
print (f"The average of your total is : {average}")
```

The Coffee Shop Inventory Manager
- The Goal: Create a dictionary pre-filled with cafe supplies and their quantities (e.g., {'coffee beans': 50, 'cups': 100, 'syrup': 15}). Create a loop that asks the user if they want to "sell" or "restock" an item. Update the dictionary values based on their input and print the current inventory levels. Prevent the user from selling more items than they have in stock!
- Concepts Tested: Accessing and updating dictionary values, checking if a key exists using the in operator, combining if/elif logic with dictionary lookups, and managing state.

```python
Coffee_shop = {
    "coffee_beans" : 20,
    "milk" : 30,
    "sugar" : 10,
}
while True:
    user_choice = input("Enter if you want to sell or restock :").lower()
    if user_choice == "sell":
        sell_item = input("Enter the items between coffee_beans, milk or sugar:").lower()
        if sell_item in Coffee_shop:
            sell_amount = int(input("Enter the amount:"))
            if sell_amount > Coffee_shop[sell_item]:
                    print ("Out of stock")
            else:
                    Coffee_shop[sell_item] -= sell_amount
                    print (f"Current stock : {Coffee_shop[sell_item]}")
        else:
            print ("Item not founded")
            break
    if user_choice == "restock":
        restock_item = input("Enter the items between coffee_beans, milk or sugar:").lower()
        if restock_item in Coffee_shop:
            restock_amount = int(input("Enter the amount:"))
            Coffee_shop[restock_item] += restock_amount
            print (f"Current stock : {Coffee_shop[restock_item]}")
        else:
            print ("Item not founded")
            break
```
